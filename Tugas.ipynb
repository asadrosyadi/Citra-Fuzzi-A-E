```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline

# Generate universe variables
x_mean = np.arange(75, 110, 1)
x_sum = np.arange(50, 105, 1)

# Generate fuzzy membership functions
mean_b = fuzz.trapmf(x_mean, [0, 75, 80, 83])
mean_d  = fuzz.trapmf(x_mean, [80, 82, 85, 87])
mean_e  = fuzz.trapmf(x_mean, [85, 87, 90, 93])
mean_c  = fuzz.trapmf(x_mean, [90, 93, 100, 105])
mean_a = fuzz.trapmf(x_mean, [97, 99, 100, 110])

sum_b = fuzz.trapmf(x_sum, [0, 50, 55, 60])
sum_e  = fuzz.trapmf(x_sum, [55, 60, 62, 64])
sum_d  = fuzz.trapmf(x_sum, [60, 62, 65, 70])
sum_c  = fuzz.trapmf(x_sum, [70, 80, 95, 97])
sum_a = fuzz.trapmf(x_sum, [95, 97, 100, 105])

A   =1.0
D  =0.8
E   =0.6
C  =0.4
B  =0.2

# Visualize these universes and membership functions
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 8))

ax0.plot(x_mean, mean_b, 'b', linewidth=1.5, label='B')
ax0.plot(x_mean, mean_d, 'g', linewidth=1.5, label='D')
ax0.plot(x_mean, mean_e, 'r', linewidth=1.5, label='E')
ax0.plot(x_mean, mean_c, 'y', linewidth=1.5, label='C')
ax0.plot(x_mean, mean_a, 'grey', linewidth=1.5, label='A')

ax0.set_title('Mean')
ax0.legend()

ax1.plot(x_sum, sum_b, 'b', linewidth=1.5, label='B')
ax1.plot(x_sum, sum_e, 'g', linewidth=1.5, label='E')
ax1.plot(x_sum, sum_d, 'r', linewidth=1.5, label='D')
ax1.plot(x_sum, sum_c, 'y', linewidth=1.5, label='C')
ax1.plot(x_sum, sum_a, 'grey', linewidth=1.5, label='A')

ax1.set_title('Sum')
ax1.legend()

ax2.plot([5, 5], [0, A], 'b', linewidth=1.5, label='A')
ax2.plot([4, 4], [0, C], 'g', linewidth=1.5, label='C')
ax2.plot([3, 3], [0, E], 'r', linewidth=1.5, label='E')
ax2.plot([2, 2], [0, D], 'y', linewidth=1.5, label='D')
ax2.plot([1, 1], [0, B], 'grey', linewidth=1.5, label='B')

ax2.set_title('Hasil')
ax2.legend()
# Turn off top/right axes
for ax in (ax0, ax1):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python

# Fuzzification
in_1 = []
in_1.append(fuzz.interp_membership(x_mean, mean_b, in_mean))
in_1.append(fuzz.interp_membership(x_mean, mean_d, in_mean))
in_1.append(fuzz.interp_membership(x_mean, mean_e, in_mean))
in_1.append(fuzz.interp_membership(x_mean, mean_c, in_mean))
in_1.append(fuzz.interp_membership(x_mean, mean_a, in_mean))

in_2 = []
in_2.append(fuzz.interp_membership(x_sum, sum_b, in_sum))
in_2.append(fuzz.interp_membership(x_sum, sum_e, in_sum))
in_2.append(fuzz.interp_membership(x_sum, sum_d, in_sum))
in_2.append(fuzz.interp_membership(x_sum, sum_c, in_sum))
in_2.append(fuzz.interp_membership(x_sum, sum_a, in_sum))

print("Derajat Keanggotaan Mean:")
if in_1[0]>0 :
    print("B :"+ str(in_1[0]))
if in_1[1]>0 :
    print("E :"+ str(in_1[1]))
if in_1[2]>0 :
    print("C :"+ str(in_1[2]))
if in_1[3]>0 :
    print("D :"+ str(in_1[3]))
if in_1[4]>0 :
    print("A :"+ str(in_1[4]))

print("")
print("Derajat Keanggotaan Sum:")
if in_2[0]>0 :
    print("B :"+ str(in_2[0]))
if in_2[1]>0 :
    print("E :"+ str(in_2[1]))
if in_2[2]>0 :
    print("C :"+ str(in_2[2]))
if in_2[3]>0 :
    print("D :"+ str(in_2[3]))
if in_2[4]>0 :
    print("A :"+ str(in_2[4]))
```

```python
## Inference dan Defazzification dengan Centroid Sugeno: 
#Penyebut
rul = []
for i in range(5):
    for j in range(5):
        rule=fuzz.relation_min(in_1[i], in_2[j])
        rul.append(rule)
        #print(rule)
penyebut=np.sum(rul)
#Pembilang
rul = []
for i in range(5):
    for j in range(5):
        rule=fuzz.relation_min(in_1[i], in_2[j])
        rulxx= rule*FAM[i][j]
        rul.append(rulxx)
        #print(rulxx)
pembilang=np.sum(rul)
hasil = pembilang/penyebut

## Logic untuk hasil
print("Keputusan :"+ str(hasil))
if hasil >=0.2 and hasil < 0.4 :
    za = (abs(hasil-0.2)/(0.4-0.2))*100
    zb = (abs(hasil-0.4)/(0.4-0.2))*100
    print("B :"+ '{:2.2f}'.format(zb)+" %")
    print("D  :"+ '{:2.2f}'.format(za)+" %")
if hasil>=0.4 and hasil< 0.43:
    za = (abs(hasil-0.4)/(0.6-0.4))*100
    zb = (abs(hasil-0.6)/(0.6-0.4))*100
    print("C :"+ '{:2.2f}'.format(zb)+" %")
    print("B :"+ '{:2.2f}'.format(za)+" %")    
if hasil>=0.43 and hasil< 0.6:
    za = (abs(hasil-0.4)/(0.6-0.4))*100
    zb = (abs(hasil-0.6)/(0.6-0.4))*100
    print("E :"+ '{:2.2f}'.format(zb)+" %")
    print("B :"+ '{:2.2f}'.format(za)+" %")
if hasil>=0.6 and hasil< 0.8:
    za = (abs(hasil-0.6)/(0.8-0.6))*100
    zb = (abs(hasil-0.8)/(0.8-0.6))*100
    print("D :"+ '{:2.2f}'.format(zb)+" %")
    print("C :"+ '{:2.2f}'.format(za)+" %")
if hasil>=0.9:
    za = (abs(hasil-0.8)/(1.0-0.8))*100
    print("A :"+ '{:2.2f}'.format(za)+" %")

```
